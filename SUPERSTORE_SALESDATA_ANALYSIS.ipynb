# EDA ON OUR DATASET FOR ANALYSIS

```python
cat_profit=df.groupby('Category')['Profit'].sum()
```

```python
# plot this column
cat_profit.plot(kind='bar')
plt.title('Profit By Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
```

CONCLUSION :- FROM THE ABOVE ANALYSIS WE CAN CLEARLY SEE THAT THE TECHNOLOGY PRODUCTS HAVE THE HIGHEST PROFIT EARNING.

```python
region_sales=df.groupby('Region')['Sales'].sum()
```

```python
#visualise this through bar
region_sales.plot(kind='bar')
plt.title('Total Sales By Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

CONCLUSION :- CENTRAL  REGION HAS THE HIGHEST SALES IN COMPARED TO OTHER REGIONS.

```python
df['Order Month']=pd.DatetimeIndex(df['Order Date']).month
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Order Month']
```

```python
month_sales=df.groupby('Order Month')['Sales'].sum()
```

```python
month_sales.plot(kind='line')
plt.title('Total Sales By Month ')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

CONCLUSION :- SALES ARE HIGHER DURING THE WINTER SEASON OR WE CAN SAY THAT SALES ARE HIGHER DURING THE END OF THE YEAR.

```python
total_orders_by_shipping_mode =df.groupby('Ship Mode').size()
returned_orders_by_shipping_mode=df[df['Profit']<0].groupby('Ship Mode').size()
returned_percentage_by_shipping_mode=(returned_orders_by_shipping_mode/total_orders_by_shipping_mode)*100
returned_percentage_by_shipping_mode
```

```python
returned_percentage_by_shipping_mode.plot(kind='bar')
plt.title('returned_percentage_by_shipping_mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Returned percentage')
plt.show()
```

CONCLUSION :- ORDERS WITH SAME DAY SHIPPING HAVE THE LOWEST RATE OF RETURNED PRODUCT.

```python
df['Order Day']=pd.DatetimeIndex(df['Order Date']).day_name()
day_profit=df.groupby('Order Day')['Profit'].sum()
day_profit
```

```python
day_profit.plot(kind='bar')
plt.title('Total_profit_bythe_day_ofthe_week')
plt.xlabel('Day_of_the_week')
plt.ylabel('Total_profit')
plt.show()
```

CONCLUSION :- COMPANY'S PROFIT IS HIGHER IN WEEKDAYS AS COMPARED TO WEEKENDS
