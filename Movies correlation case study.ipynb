```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# load the dataset
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#looking for the null values
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df=df.drop(columns='year',axis=1)
```

```python
df.head()
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
df.head()
```

```python
df.columns
```

# budget high correlation
# company high correlation


```python
#scatterplot with budget vs gross

plt.scatter(df['budget'],df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Budget for movie')
plt.ylabel('Gross earnings')
plt.show
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'alpha': 1.0},line_kws={'color':'red'})
plt.title('Budget vs Gross earnings')
```

```python
df.corr()
```

high correlation between budget and gross

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation metric for numeric features')
```

Budget and Votes are highly correlated with the Gross.
